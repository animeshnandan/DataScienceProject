```python
from google.colab import drive
import os
import pandas as pd

drive.mount('/content/drive')
PATH = 'drive/MyDrive/Fall 2023/ENPM808W/ENPM808W Project Dataset/train.csv'

data = pd.read_csv(PATH)
data.head()
```

```python
#missing values and datatypes
missing_values = data.isnull().sum()
data_info = data.info()

#duplicate rows
duplicate_rows = data.duplicated().sum()
missing_values, duplicate_rows, data_info
```

```python
#duplicate rows
data_cleaned = data.drop_duplicates()
data_cleaned.shape
```

```python
#statistics
descriptive_stats = data_cleaned.describe()
descriptive_stats.transpose()
```

```python
import matplotlib.pyplot as plt

features_to_plot = ['number_of_elements', 'mean_atomic_mass', 'entropy_atomic_mass', 'wtd_std_fie', 'critical_temp']

plt.figure(figsize=(15, 10))

for i, feature in enumerate(features_to_plot, 1):
    plt.subplot(2, 3, i)
    data_cleaned[feature].hist(bins=20)
    plt.title(feature)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

correlation_features = features_to_plot + ['wtd_mean_atomic_mass', 'std_atomic_mass', 'entropy_fie']
correlation_matrix = data_cleaned[correlation_features].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#scatter plots against critical_temp
plt.figure(figsize=(20, 10))

for i, feature in enumerate(features_to_plot[:-1], 1):
    plt.subplot(2, 2, i)
    plt.scatter(data_cleaned[feature], data_cleaned['critical_temp'], alpha=0.5)
    plt.title(f"{feature} vs Critical Temperature")
    plt.xlabel(feature)
    plt.ylabel('Critical Temperature (K)')

plt.tight_layout()
plt.show()
```
